```python
query = "SELECT Crop_Type,Irrigation_Type,Soil_Type,Season FROM (select * from insights_Df where Crop_Type in ('Wheat','Maize','Barley', 'Rice','Sugarcane','Soybean')) AS Stagging group by Crop_Type"
Irrigation = ps.sqldf(query, locals())
pd.set_option('display.width', 1000)
Irrigation
```

```python
query = "SELECT Crop,Year,Country,Region, Average_Temperature_C, Total_Precipitation_mm, CO2_Emissions_MT, Crop_Yield_MT_per_HA, Extreme_Weather_Events, [Irrigation_Access_%], Pesticide_Use_KG_per_HA, Fertilizer_Use_KG_per_HA, Soil_Health_Index, Adaptation_Strategies, Economic_Impact_Million_USD, Irrigation_Type, Soil_Type, Season FROM (Select * from filtered_climatechange_Df where Crop in ('Wheat','Maize','Barley', 'Rice','Sugarcane','Soybean')) AS A Left JOIN Irrigation  AS B ON TRIM(A.Crop) = TRIM(B.Crop_Type)"
CombinedDf = ps.sqldf(query, locals())
pd.set_option('display.width', 1000)
CombinedDf
```
